```python
import random
```

```python
pospath = "../data/positive_labeled.txt"
negpath= "../data/negative_labeled.txt"
trainpath = "../data/train.xml.seg"
testpath = "../data/test.xml.seg"
```

```python
class SampleMaker:
    def __init__(self,pospath,negpath,trainpath,testpath):
        self.pospath = pospath
        self.negpath = negpath
        self.trainpath = trainpath
        self.testpath = testpath
        
    def readtxt(self, path):
        txtfile = open(path,"r",encoding='utf-8',errors='ignore')
        lines = txtfile.readlines()
                
        flag = 0
        self.review_list = []
        self.term_list = []
        self.aspect_list = []
        
        review_temp = ""
        term_temp = []
        aspect_temp = []
        
        cnt = 0
        for line in lines:
            if "EOF" in line:
                self.aspect_list.append(aspect_temp)
                break
            if "*****O" in line:
                flag = 1
                if cnt != 0:
                    self.aspect_list.append(aspect_temp)
                    aspect_temp = []
                cnt += 1
                continue
            elif "***T" in line:
                flag = 2
                self.review_list.append(review_temp)
                review_temp = ""
                continue
            elif "***A" in line:
                if flag == 1:
                    self.review_list.append(review_temp)
                    review_temp = ""
                    
                flag = 3
                self.term_list.append(term_temp)
                term_temp = []
                continue
              
            if flag == 1:
                review_temp = line.replace('\n','')
                continue
            if flag == 2:
                term_temp.append(line.replace('\n',''))
                continue
            if flag == 3:
                aspect_temp.append(line.replace('\n',''))
                continue
            
        #print(self.review_list)
        #print('/////////////////////////////////')
        #print(self.term_list)
        #print('/////////////////////////////////')
        #print(self.aspect_list)

    def sample_help(self):
        final_list_tmp = []
        
        for i in range(len(self.review_list)):
            
            num = len(self.term_list[i])
            if num == 0:
                continue
            elif num == 2:
                #original = self.review_list[i]
                #original.split(self.term_list[i][0])
#                 final_list.append(self.review_list[i].replace(self.term_list[i][0], '$T$', 1))
#                 final_list.append(self.term_list[i][0])
#                 final_list.append(self.term_list[i][1])
                tmp = (self.review_list[i].replace(self.term_list[i][0], '$T$', 1), 
                      self.term_list[i][0],
                      self.term_list[i][1])
                final_list_tmp.append(tmp)
            if num == 4:
#                 final_list.append(self.review_list[i].replace(self.term_list[i][0], '$T$', 1))
#                 final_list.append(self.term_list[i][0])
#                 final_list.append(self.term_list[i][1])
#                 final_list.append(self.review_list[i].replace(self.term_list[i][2], '$T$', 1))
#                 final_list.append(self.term_list[i][2])
#                 final_list.append(self.term_list[i][3])
                tmp = (self.review_list[i].replace(self.term_list[i][0], '$T$', 1), 
                      self.term_list[i][0],
                      self.term_list[i][1])
                final_list_tmp.append(tmp)
        
                tmp = (self.review_list[i].replace(self.term_list[i][2], '$T$', 1), 
                      self.term_list[i][2],
                      self.term_list[i][3])
                final_list_tmp.append(tmp)
                
        return final_list_tmp
            
    def sample(self):
        final_train = []
        final_test = []
        final_list_train = []
        final_list_test = []
        
        self.readtxt(self.pospath)
        final_list_tmp = self.sample_help()
        
        final_list_train += final_list_tmp[:int(len(final_list_tmp)*0.7)]
        final_list_test += final_list_tmp[int(len(final_list_tmp)*0.7):]
        
        self.readtxt(self.negpath)
        final_list_tmp = self.sample_help()
        final_list_train += final_list_tmp[:int(len(final_list_tmp)*0.7)]
        final_list_test += final_list_tmp[int(len(final_list_tmp)*0.7):]
        
        print(len(final_list_train),len(final_list_test))
                
        
        random.shuffle(final_list_train)
        random.shuffle(final_list_test)
        
        for i in final_list_train:
            tp1,tp2,tp3 = i
            final_train.append(tp1)
            final_train.append(tp2)
            final_train.append(tp3)
            
        for i in final_list_test:
            tp1,tp2,tp3 = i
            final_test.append(tp1)
            final_test.append(tp2)
            final_test.append(tp3)
            

        f = open(self.trainpath, 'w')
        for i in range(len(final_train)):
            f.write(final_train[i])
            f.write('\n')
        f.close()
        
        
        f = open(self.testpath, 'w')
        for i in range(len(final_test)):
            f.write(final_test[i])
            f.write('\n')
        f.close()
```

```python
read = SampleMaker(pospath,negpath,trainpath,testpath)
# print(len(read.review_list),len(read.term_list),len(read.aspect_list))
read.sample()
```
